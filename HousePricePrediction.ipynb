# **House Price Prediction**

Importing all the required libraries

> Pandas : To read the dataset and display in readable format.

> Sklearn : It contains all the Machine Learning Models.

> Matplotlib : It is Used to plot graphs.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv('kc_house_data.csv')
house_data.dropna().reset_index(drop=True)
```

```python
house_data.head(5).describe
```

```python
X = house_data.drop(['price','id','date'],axis = 1)
Y = house_data['price']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y)
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
prediction = model.predict(xtest)
```

```python
mse = mean_squared_error(ytest, prediction)
print(f"Mean Squared Error: {mse}")
```

```python
plt.scatter(ytest, prediction, alpha=0.5)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs. Predicted Housing Prices')
plt.show()
```

```python
coefficients = model.coef_

low_coef_indices = [index for index, coef in enumerate(coefficients) if -2000 <= coef <= 2000]

xtrain_filtered = xtrain.drop(xtrain.columns[low_coef_indices],axis = 1)
xtest_filtered = xtest.drop(xtest.columns[low_coef_indices],axis = 1)
```

```python
model.fit(xtrain_filtered,ytrain)
new_prediction = model.predict(xtest_filtered)
```

```python
plt.scatter(ytest, new_prediction, alpha=0.5)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs. Predicted Housing Prices')
plt.show()
```
